Примечание\
Напомним, что дерево — это связный граф на n вершинах, состоящий из n - 1 ребра.\
Подвешенное дерево — это дерево с особой выделенной вершиной, корнем. В подвешенном дереве из любых двух вершин, соединённых ребром, одна называется предком (та из них, что ближе к корню), а оставшаяся — ребёнком. Вершина называется листом, если у неё нет детей.

```python
# поиск в глубину вызываем рекурсию, которая движется по веткам и считает
def DFS(graph, vert, vertcat, count, catcount, k):
    if vertcat[vert] == 1:
        if catcount + 1 <= k:
            catcount += 1
            for i in graph[vert]:
                count = DFS(graph, i, vertcat, count, catcount, k)
                
    else:
        count += 1
        for i in graph[vert]:
                count = DFS(graph, i, vertcat, count, catcount, k)
    return count
            
n,k=map(int,input().split())
vertcat=list(map(int,input().split()))
graph = [0]*n
for i in range(n):
    graph[i] = []
for i in [0]*(n-1):
    a, b = map(int,input().split())
    graph[a-1].append(b-1)
# print(graph)
# graph = [[1,2],[3,4],[5,6],[],[],[],[]]
# vertcat = [1,0,1,1,0,0,0]
count = DFS(graph,0,vertcat,0,0,1)
print("Количество ресторанов:",count)
```

```python
# число делится на 8, если 3 последние цифры числа делится на 8
def num(num):
    num = str(num)
    for i in range(len(num)-2):
        for j in range(i+1,len(num)-1):
            for k in range(j+1,len(num)):
                if int(num[i]+num[j]+num[k]) % 8 == 0:
                    print('YES')
                    print(int(num[i]+num[j]+num[k]))
                    return
    print('NO')
num(input())
    
    
```

Друг чувствует себя бедным, если в компании есть кто-то, у кого денег хотя бы на d единиц больше, чем у него. Также Кефа хочет, чтобы суммарная степень дружбы членов компании была максимальной. Помогите ему пригласить оптимальную компанию!

Входные данные

В последующих n строках даны описания друзей Кефы, в (i + 1)-й строке содержится описание i-го друга вида mi, si (0 ≤ mi, si ≤ 109) — количество денег и степень дружбы с Кефой соответственно.

Примечание\
В первом тесте из условия выгоднее всего сформировать компанию только из второго друга. При всех других вариантах суммарная степень дружбы будет меньше.

Во втором тесте из условия мы можем взять всех друзей.

```python
# ищем подстроки друзей, у которых деньги не превосходят d, 
# выбираем максимальную подстроку друзей с наибольшей суммой дружбы

import math
# n, d = list(map(int,input().split()))
# friend = [[*map(int,input().split())] for i in range(n)]

n,d=4,5
friend = [[0, 100], [75, 1], [75, 5], [150, 20]]

n,d=5,100
friend = [[0,7],[11,32],[99,10],[46,8],[87,54]]

friend.sort() #сортируем по возрастанию денег
mxd = friend[0][1]

for i in range(n): # ищем подстроки друзей, у которых деньги не превосходят d, 
    j = i+1
    temp = friend[i][1]
    while j<n and abs(friend[i][0] - friend[j][0]) < d:
        temp += friend[j][1]
        j += 1
    i = j 
    mxd = max(mxd,temp) # выбираем максимальную подстроку друзей с наибольшей суммой дружбы
print(mxd)
```

Пусть этот отрезок равен [𝑙;𝑟]. Если 𝑟−𝑙+1 нечетно (не делится на 2), то присваивается 𝑎[𝑙+𝑟2]:=𝑖 (где 𝑖 — номер текущего действия), иначе (если 𝑟−𝑙+1 четно) присваивается 𝑎[𝑙+𝑟−12]:=𝑖.\
Рассмотрим массив 𝑎 длины 5 (изачально 𝑎=[0,0,0,0,0]). Тогда он меняется следующим образом:

Сначала мы выбираем отрезок [1;5] и присваиваем 𝑎[3]:=1, таким образом 𝑎 становится равен [0,0,1,0,0];\
затем мы выбираем отрезок [1;2] и присваиваем 𝑎[1]:=2, таким образом 𝑎 становится равен [2,0,1,0,0];\
затем мы выбираем отрезок [4;5] и присваиваем 𝑎[4]:=3, таким образом 𝑎 становится равен [2,0,1,3,0];\
затем мы выбираем отрезок [2;2] и присваиваем 𝑎[2]:=4, таким образом 𝑎 становится равен [2,4,1,3,0];\
и наконец мы выбираем отрезок [5;5] и присваиваем 𝑎[5]:=5, таким образом 𝑎 становится равен [2,4,1,3,5].\
Ваша задача — найти массив 𝑎 длины 𝑛 после выполнения всех 𝑛 действий. Заметьте, что ответ существует и единственен.

Вам необходимо ответить на 𝑡 независимых наборов тестовых данных.

Входные данные\
Первая строка входных данных содержит одно целое число 𝑡 (1≤𝑡≤104) — количество наборов тестовых данных. Затем следуют 𝑡 наборов тестовых данных.

Единственная строка набора тестовых данных содержит одно целое число 𝑛 (1≤𝑛≤2⋅105) — длину 𝑎.

Гарантируется, что сумма 𝑛 по всем наборам тестовых данных не превосходит 2⋅105 (∑𝑛≤2⋅105).

Выходные данные\
Для каждого набора тестовых данных выведите ответ — массив 𝑎 длины 𝑛 после выполнения 𝑛 действий, описанных в условии задачи. Заметьте, что ответ существует и единственен.

```python
# Добавляем в кучу отрезки, которые мы будем рассматривать(благодаря функции
# tuplmk, у которого первый элемент показывает место в куче, по этому 
# элементу сортируется куча, повторяем операцию n раз, чтобы заполнить 
# наш итоговый массив)

import heapq
def fk(a,b):
    return (a-1-b, a, b)
t = int(input())
for pset in range(t):
    n = int(input())
    h = []
    ans = [0 for i in range(n)]
    heapq.heappush(h,fk(1,n))

    for el in range(1,n+1):
        seq = heapq.heappop(h)
        ind = (seq[1]+seq[2])//2
        ans[ind-1] = el
        heapq.heappush(h,fk(seq[1], ind-1))
        heapq.heappush(h,fk(ind+1, seq[2]))
    print(" ".join([str(i) for i in ans]))
```

Входные данные

Следующая строка содержит 𝑛 целых чисел 𝑎𝑖 (1≤𝑎𝑖≤2⋅105), разделенных пробелами — элементы массива.

Каждая из следующих 𝑞 строк содержит два целых числа 𝑙𝑖 и 𝑟𝑖 (1≤𝑙𝑖≤𝑟𝑖≤𝑛), разделенных пробелом — 𝑖-й запрос.

Выходные данные
В единственной строке выведите целое число — максимальную сумму ответов на запросы после перемешивания элементов массива.

Пожалуйста, не используйте спецификатор %lld для чтения или записи 64-х битовых чисел на С++. Рекомендуется использовать потоки cin, cout или спецификатор %I64d.

```python
# считаем максимальное количество запросов каждого отрезка,
# далее сортируем этот словарь, у которого ключ - это номер последовательности,
# а значение - это количество запросов именно этого номера
# далее в новый массив добавляем максимальный элемент(мы считываем словарь,
# который отсортировали), далее отсортировываем
n,m = map(int,input().split())
h = [*map(int,input().split())]
hsum = [0]*n

numCount = dict()
seq = []

for _ in [0]*m:
    a,b = map(int,input().split())
    seq.append([a,b])
    for i in range(a,b+1):
        if i in numCount:
            numCount[i] += 1
        else:
            numCount[i] = 1

numCount={k: v for k, v in sorted(numCount.items(), key=lambda item: item[1])}
u = reversed(numCount)
ans = [0]*n

for i in u:
    ans[i-1] = h.pop(h.index(max(h)))


hsum[0] = ans[0]

for i in range(1,n):
    hsum[i] = ans[i] + hsum[i-1]
hsum=[0]+hsum
res = 0
for i in seq:
    res += hsum[i[1]]-hsum[i[0]-1]
print('Сумма:',res)
```
