### Бинарный поиск

```python
def binary_search(mas, x):
    lower_bound = 0;
    upper_bound = len(mas) - 1
    center = (lower_bound + upper_bound) // 2
    
    while (lower_bound <= upper_bound):
        center = (lower_bound + upper_bound) // 2
        if (mas[center] == x):
            return center
        elif (mas[center] < x):
            lower_bound = center + 1
        elif (mas[center] > x):
            upper_bound = center - 1
            
    return -1
```

### Фибоначчиев поиск

```python
def fibonachi(n):
    f0 = 0
    f1 = 1
    f = 0
    k = 1
    while k < n:
        f = f0 + f1
        f0 = f1
        f1 = f
        k += 1
    return f

def searchFK(n):
    f0 = 0
    f1 = 1
    f = 0
    k = 1
    while f <= n + 1:
        f = f0 + f1
        f0 = f1
        f1 = f
        k += 1
    return k-1

def fib_search(mas, el):
    n = len(mas)
    k = searchFK(n + 1)
    #print(k)
    m = fibonachi(k + 1) - (n + 1)
    #print(m)
    i = fibonachi(k) - m
    #print(i)
    p = fibonachi(k - 1)
    #print(p)
    q = fibonachi(k - 2)
    #print(q)
    while k != el:
        if i < 0:
            if p == 1:
                return -1
            i = i + q
            p = p - q
            q = q - p
            continue

        if i >= n:
            if q == 0:
                return -1
            i = i - 1
            temp = p
            p = q
            q = temp - q
            # p, q = q, p - q
            continue

        if el < mas[i]:
            if q == 0:
                return -1
            i = i - q
            temp = p
            p = q
            q = temp - q
            #p, q = q, p - q
            continue
        if el > mas[i]:
            if p == 1:
                return -1
            i = i + q
            p = p - q
            q = q - p
            continue
        if el == mas[i]:
            return i
```

### Интерполяционный поиск

```python
def interpolation_search(mas, el):
    low = 0
    high = len(mas) - 1
    midle = 0
    while low <= high and high < len(mas) and low >= 0:
        midle = low + int(((high-low) / (mas[high]-mas[low]))*(el-mas[low]))
        if (midle > high):
            return -1
        if (mas[midle] == el):
            return midle
        elif mas[midle] < el:
            low = midle + 1

        else:
            high = midle - 1
    return -1
```

### Рехэширование цепочкой

```python
class RehCep:
    def __init__(self, n):
        self.n = n
        self.lst = [[None]] * n

        self.count = 0

    def ind(self, el):
        return (len(el) + ord(el[0])) % self.n

    def add(self, el):
        if self.count == self.n:
            print('Массив заполнен')
            return

        k = self.ind(el)
        if self.lst[k][0] is None:
            self.lst[k] = [el]
        else:
            self.lst[k].append(el)

        self.count += 1

    def searchEl_C(self, el):
        k = self.ind(el)
        i = 0
        if self.lst[k][i] == None:
            return -1, -1
        else:
            p = len(self.lst[k])
            while i >= len(self.lst[k]):
                if el == self.lst[k][i]:
                    break
                else:
                    i += 1

        if i < len(self.lst[k]) and el == self.lst[k][i]:
            return k, i
        else:
            return -1,-1
```
