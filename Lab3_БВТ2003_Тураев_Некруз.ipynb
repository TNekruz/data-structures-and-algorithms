```python
def pref(str):
    pref = [0]*len(str)
    beg = 0
    end = 1
    n = len(str)
    while beg < n and end < n:
        if str[beg] != str[end]:
            if beg == 0:
                pref[end] = 0
                end += 1
            else:
                beg = pref[beg-1]
        else:
            pref[end] = beg+1
            beg += 1
            end += 1
    return pref

def KMP(str, obr):
    n = len(str)
    m = len(obr)
    d = 0 # указывает на символ образца
    i = 0 # указывает на символ строки
    k = pref(obr)
    while i < n:
        if d == m-1 and obr[d] == str[i]:
            return True
            break
        if str[i] == obr[d]:
            d += 1
            i += 1
        elif d != 0:
            d = k[d-1]
        else:
            i += 1
    if i == n:
        return False
```

```python
def CheckP(mas): #Проверка на решаемость
    k = 0
    for i in range(len(mas)):
        for j in range(i+1,len(mas)):
            if mas[i] > mas[j] and mas[j] != 0:
                k += 1
    z = mas.index(0) // 4 + 1
    k = z + k
    #print(k,z)
    if (k % 2 == 0):
        return True
    else:
        return False


def zeroL(iz,jz,im,jm,matr): #Ставим 0 слева от num
    # 0 ставим слева от num
    if iz == im and jz == jm-1:
        return iz,jz,matr

    while iz < im: #Опускаем
        if jz == jm: #Если на одном столбце опускаемся до num и встаём справа
            while iz+1 < im:
                matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
                print(matr[iz][jz], end=' ')
                iz += 1
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz -= 1

        matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
        print(matr[iz][jz], end=' ')
        iz += 1


    while iz > im:
        if jz == jm:
            while iz-1 > im:
                matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
                print(matr[iz][jz], end=' ')
                iz -= 1
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz -= 1

        matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
        print(matr[iz][jz], end=' ')
        iz -= 1

    if jz > jm: #если справа от num, то опускаем и двигаем влево, после вверх
        f = True
        if iz == 3 and iz == im: # если на одной строке и в самом низу
            f = False
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1
        elif iz != 3 and iz == im:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        while jz+1 > jm:
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz -= 1
        if not f:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        else:
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1
    while jz + 1 < jm:
        matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
        print(matr[iz][jz], end=' ')
        jz += 1
    return iz, jz, matr






def zeroR(iz,jz,im,jm,matr):
    # 0 ставим справа от num
    if iz == im and jz == jm+1:
        return iz,jz,matr
    while iz < im: #опускаем
        if jz == jm:
            while iz+1 < im:
                matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
                print(matr[iz][jz], end=' ')
                iz += 1
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1

        matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
        print(matr[iz][jz], end=' ')
        iz += 1


    while iz > im: #поднимаем
        if jz == jm:
            while iz-1 > im:
                matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
                print(matr[iz][jz], end=' ')
                iz -= 1
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1

        matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
        print(matr[iz][jz], end=' ')
        iz -= 1

    if jz < jm: #если слева от num, то опускаем и двигаем влево, после вверх
        f = True
        if iz == 3 and iz == im: # Если на одной строке
            f = False
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1
        elif iz != 3 and iz == im:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        while jz <= jm:
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1
        if not f:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        else:
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1

    while jz-1 > jm:
        matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
        print(matr[iz][jz], end=' ')
        jz -= 1

    return iz, jz, matr








def numL(iz,jz,im,jm,l,matr):
    #у нас 0 должен стоять слева от num
    while jm > l:  # num left
        f = True
        if jz < jm:
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz, jm = jm, jz
        if jm == l:
            return iz,jz,im,jm,matr

        if iz == 3:
            f = False
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1
        else:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1

        matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
        print(matr[iz][jz], end=' ')
        jz -= 1

        matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
        print(matr[iz][jz], end=' ')
        jz -= 1

        if not f:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        else:
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1

        matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
        print(matr[iz][jz], end=' ')
        jz, jm = jm, jz
    return iz,jz,im,jm,matr

def numR(iz,jz,im,jm,l,matr):
    #у нас 0 должен стоять справа от num
    while jm < l:  # num right
        f = True
        if jz > jm:
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz, jm = jm, jz

        if jm == l:
            return iz,jz,im,jm,matr

        if iz == 3:
            f = False
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1
        else:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1

        i = 0
        while jz <= jm and jz != 3 and i < 2:
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1
            i+=1

        if jm == 3:
            continue
        if not f:
            matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
            print(matr[iz][jz], end=' ')
            iz += 1
        else:
            matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
            print(matr[iz][jz], end=' ')
            iz -= 1

        matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
        print(matr[iz][jz], end=' ')
        jz, jm = jm, jz
    return iz,jz,im,jm,matr

def numUp(iz,jz,im,jm,k,matr):
    # 0 должен быть справа или слева от num
    while im > k:
        f = True
        matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
        print(matr[iz][jz], end=' ')
        iz -= 1

        if jm < jz:
            f = False
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz -= 1
        else:
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1

        matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
        print(matr[iz][jz], end=' ')
        iz, im = im, iz

        if (not f):
            matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
            print(matr[iz][jz], end=' ')
            jz += 1
        else:
            matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
            print(matr[iz][jz], end=' ')
            jz -= 1

        matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
        print(matr[iz][jz], end=' ')
        iz -= 1
    return iz,jz,im,jm,matr



def c125(num,k,l, mas):
    matr = [mas[0:4],mas[4:8],mas[8:12],mas[12:16]]

    #k = num // 4
    #l = num % 4 -1
    ind = mas.index(num)
    zero = mas.index(0)
    im = ind // 4
    jm = ind % 4
    iz = zero // 4
    jz = zero % 4

    if jm > l:
        #0 ставим слева от num
        iz, jz, matr = zeroL(iz,jz,im,jm,matr)

        # Двигаем num влево
        iz, jz, im, jm, matr = numL(iz, jz, im, jm, l, matr) # num left
    elif jm < l:
        # 0 ставим справа от num
        iz, jz, matr = zeroR(iz, jz, im, jm, matr)

        # Двигаем num вправо
        iz, jz, im, jm, matr = numR(iz, jz, im, jm, l, matr)



    # Поднимаем num вверх
    if im > k and l < 3:
        # 0 ставим справа от num
        iz, jz, matr = zeroR(iz, jz, im, jm, matr)
        # Поднимаем num вверх
        iz, jz, im, jm, matr = numUp(iz,jz,im,jm,k,matr)
    elif im > k and l == 3:
        # 0 ставим слева от num
        iz, jz, matr = zeroL(iz, jz, im, jm, matr)
        # Поднимаем num вверх
        iz, jz, im, jm, matr = numUp(iz, jz, im, jm, k, matr)
    g = 0
    for i in matr:
        for j in i:
            mas[g] = j
            g+=1
    print()
    return mas

def c34(num,mas):
    matr = [mas[0:4], mas[4:8], mas[8:12], mas[12:16]]
    num2 = num-1
    # k = num // 4
    # l = num % 4 -1
    ind = mas.index(num)
    ind2 = mas.index(num2)
    zero = mas.index(0)
    im = ind // 4
    jm = ind % 4
    im2 = ind2 // 4
    jm2 = ind2 % 4
    iz = zero // 4
    jz = zero % 4
    #print(matr)

    iz, jz, matr = zeroR(iz, jz, im2, jm2, matr) #0 right

    while iz > im:
        matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
        print(matr[iz][jz], end=' ')
        iz -= 1
    #print(matr)
    matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
    print(matr[iz][jz], end=' ')
    jz -= 1

    matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
    print(matr[iz][jz], end=' ')
    iz += 1

    print()
    g = 0;
    for i in matr:
        for j in i:
            mas[g] = j
            g+=1

    return mas

def c913(num, mas):
    matr = [mas[0:4], mas[4:8], mas[8:12], mas[12:16]]

    ind = mas.index(num)
    zero = mas.index(0)
    #im = ind // 4
    jm = ind % 4
    iz = zero // 4
    jz = zero % 4

    while iz < 3:
        matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
        print(matr[iz][jz], end=' ')
        iz += 1
    while jz > jm:
        matr[iz][jz], matr[iz][jz - 1] = matr[iz][jz - 1], matr[iz][jz]  # 0 left
        print(matr[iz][jz], end=' ')
        jz -= 1

    matr[iz - 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz - 1][jz]  # 0 up
    print(matr[iz][jz], end=' ')
    iz -= 1

    matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
    print(matr[iz][jz], end=' ')
    jz += 1

    g = 0
    for i in matr:
        for j in i:
            mas[g] = j
            g += 1

    return mas

def cend(mas):
    matr = [mas[0:4], mas[4:8], mas[8:12], mas[12:16]]

    #ind = mas.index(num)
    zero = mas.index(0)
    # im = ind // 4
    #jm = ind % 4
    iz = zero // 4
    jz = zero % 4

    matr[iz + 1][jz], matr[iz][jz] = matr[iz][jz], matr[iz + 1][jz]  # 0 down
    print(matr[iz][jz], end=' ')
    iz += 1

    matr[iz][jz], matr[iz][jz + 1] = matr[iz][jz + 1], matr[iz][jz]  # 0 right
    print(matr[iz][jz], end=' ')
    jz += 1

    g = 0
    for i in matr:
        for j in i:
            mas[g] = j
            g += 1

    return mas
```
